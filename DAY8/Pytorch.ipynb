## 1. Normalization:

Taking every sample in dataset and transforming into new set of values. New values also encodes a new kind of information relative to dataset. Data Normalization is done to rescale the feature. This process is particularly useful when the features have different ranges, and you want to bring them to a common scale.
- For example we are having a dataset contaning 2 attributes weight in pounds and height in inches both belong to 2 different scales to bring them to one scale ranging between 0 & 1 we use normalization technique.


***-> Feature Scaling:***

   Feature scaling is a preprocessing technique used in machine learning to normalize or standardize the features (input variables) of a dataset before they are fed into a model.
***-> Standardization:***

Standardization is a specific term. We use Standardization technique to normalize our data. In standardization we use a formula:

***Formula:***        

***X_standardized = (X - X_mean) / X_std***

   Here we subtract values of from the mean and dividing it with std. Std is calculated by taking square of variance. standardization is about transforming data to have a mean of 0 and a standard deviation of 1.

## 2. DataLoaders:

- DataLoaders are classes or components within libraries like PyTorch that help manage the loading and processing of data. Datasets are passed through DataLoaders in order to facilitate the process of loading and preparing the data for training or evaluation of machine learning models, especially in scenarios where the dataset is too large to fit entirely into memory. 
- DataLoaders are particularly useful when working with large datasets that cannot be loaded entirely into memory at once. The DataLoader is a higher-level component that wraps around a Dataset and provides functionalities to efficiently load data in batches. It takes care of shuffling the data, creating batches, and potentially using multi-process loading to speed up the data loading process. The DataLoader class provides an iterator interface, allowing you to loop through mini-batches of data.

- ***Dataset:***

Firstly, dataset is provided to the model to train. Dataloaders split these large set of data.
- ***Batch-size:***

Dataloaders converts this large set of data into small batch because entire data may not be load into the memory.
- ***Shuffling of Data:***

Shuffling of data is necessary to break the patterns and prevent the model from relying on specific orders of data points is primarily done to enhance the generalization ability of the machine learning model. The generalization ability of a machine learning model refers to its capability to perform well on new, unseen data that it hasn't been specifically trained on. 

## 3. Epochs:

An epoch means that entire model has been feeded with entire dataset at once. This is done to several times so that our model learn about that data. We give data to model for several times so that there should be less chance of error in future.

- ***For Example:***

  We have 1000 images and we feed them at once so it will be one epoch and when we feed the same 1000 images for 50 times it will be 50 epochs.

## 4. Torchvision:

A library built to facilitate research and experimentation in Computer Vision. It provides a collection of tools and utilities for working with image and video data, making it easier to perform tasks related to computer vision using PyTorch. Torchvision provides

- ***Dataset and Dataloaders:***

  TorchVision includes a collection of popular computer vision datasets that cover a variety of tasks such as image classification, object detection, semantic segmentation, and more.
- ***Transformations:***

- ***Models:***

- ***Video Processing:***

   Its main focus is on images, TorchVision also provides utilities for working with video data.

## 5.OverFitting:

 - The curve shaped curve obtained after the training of model as ouput represents overfitting.
 - The model or line obtained after training of data is represented by a line in a 2 dimensional graph.
 - When model or line covers every datapoint placed in graph that is overfitting.
 - All the datapoints are on the line.
 - Overfitting occurs in machine learning when a model learns the training data so well that it starts to memorize it, including the noise and random fluctuations.
 - When we give extra attributes or features to train our model it leads to wrong prediction.
 

## 6. UnderFitting:

 - The line or model obtained after training the dataset as output represents underfitting.
 -  When model or line donot cover all the datapoints placed on graph then it is called underfitting.
 -   Underfitting in simple words can be described as a situation where a model is too basic to understand or represent the complexity of the data it's trying to learn from.
 -   When we give less attributes or features to train our model this is Underfitting.

## 7. Good Fit/ Robust:

- A "good fit" or "robust" model in machine learning is one that achieves a balance between underfitting and overfitting.
-  It accurately captures the underlying patterns in the training data without being overly complex or memorizing noise.
-  When we give that attributes or features that are necessary to train a model will lead to good prediction and us Good Fit.

## Transfer learning:
 - Transfer learning is a powerful technique for working with PyTorch model input shape.
 -  Transfer learning involves using a pre-trained model as a starting point for your own model. By using a pre-trained model, you can leverage the knowledge that has already been learned by the model and fine-tune it for your specific task.

## Input and Output Shape:

 - ***Input Shape*** is usually given in the form of tensors. Tensors is a multi-dimensional array that holds numerical data. Think of the input shape as describing how you give information to a machine learning model. Imagine you have a bunch of pictures of different animals that you want the model to recognize. The input shape tells the model the size and type of each picture.

 - ***Output Shape*** the output shape is like the response you get from the model after it has looked at the pictures. Let's say you want the model to say what animal it thinks is in each picture. The output shape tells you how many options the model is considering. For example, if the model knows about cats, dogs, and birds, the output shape might tell you "I'm going to tell you my guess for each picture using three numbers because I know about these three animals."
