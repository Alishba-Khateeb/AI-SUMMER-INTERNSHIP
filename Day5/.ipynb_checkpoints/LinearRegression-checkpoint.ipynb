```python
import matplotlib.pyplot as plt
plt.plot(X,y,'bo')
plt.ylabel('HappinessScore')
plt.xlabel('GDP_PerCapita')
plt.legend(['HappinessScore'])
plt.title('GDP_PerCapita_Vs_Happiness')
plt.grid()
plt.show()
```
