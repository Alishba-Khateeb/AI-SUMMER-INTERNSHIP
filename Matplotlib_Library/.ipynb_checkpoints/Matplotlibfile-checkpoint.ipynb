```python
x=np.arange(1,11)
y1=x*2
y2=x*4
plt.subplot(1,2,1)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y1 , color='purple', linestyle='-.', linewidth=2)
plt.subplot(1,2,2)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y1 , color='black', linestyle='-.', linewidth=2)
plt.show()
```

```python
x=np.arange(1,11)
y1=x*2
y2=x*4
plt.subplot(2,1,1)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y1 , color='purple', linestyle='-.', linewidth=2)
plt.subplot(2,1,2)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y1 , color='black', linestyle='-.', linewidth=2)
plt.show()
```

```python
x = ["BOB", "SAM", "CLERK"]
y = [90, 60, 43]
plt.bar(x, y)
plt.show()
```

```python
student={"BOB":90, "SAM":65, "CLERK":43}  #Another way to plot a graph
```

```python
names= list(student.keys())
values= list(student.values())
```

```python
plt.bar(names, values)
plt.show()
```

```python
plt.bar(names, values)
plt.title("BAR PLOT")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
plt.barh(names, values, color='purple')#barh for horizontal graph
plt.title("BAR PLOT")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.show()
```

```python
fruit= ["apple","orange","mango","peach"]
value= [30,70,40,25]
```

```python
plt.pie( value,labels=fruit)
plt.show()
```

```python
#autopct to add the percentage and 0.1 means decimal to 1 place
plt.pie( value,labels=fruit, autopct= '%0.1f%%')
plt.show()
```

```python
#adding colors
#autopct to add the percentage and 0.1 means decimal to 1 place
plt.pie( value,labels=fruit, autopct= '%0.1f%%', 
        colors=["red","orange", "yellow", "pink" ])
plt.show()
```
